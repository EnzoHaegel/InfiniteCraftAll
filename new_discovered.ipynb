```python
import json

def print_new_items(file_path):
    with open(file_path, 'r') as file:
        items_data = json.load(file)
    
    for item in items_data['items']:
        if item['isNew']:
            print(item['emoji'], item['result'])

file_path = './items.json'
print_new_items(file_path)
```
